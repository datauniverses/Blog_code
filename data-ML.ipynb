```python
import pandas as pd
```

## Filtering Non-Employment Data for Middle Eastern Countries

```python
middle_east_non_emp = pd.read_excel(r"data\middle east unemployment rates\API_SL.UEM.TOTL.ZS_DS2_en_excel_v2_5358380.xls")
len(middle_east_non_emp)
pd.set_option('display.max_rows', None)
```

```python
middle_east_non_emp.iloc[:, :]
```

```python
countries_list = ['Country Name', 'Bahrain', 'Iran', 'Iraq', 'Israel', 'Jordan', 'Kuwait', 'Lebanon', 'Oman', 'Qatar', 'Saudi Arabia', 'Syria', 'United Arab Emirates', 'Syrian Arab Republic', 'Iran, Islamic Rep.','Yemen, Rep.' ,'West Bank and Gaza']

middle_east_countries = middle_east_non_emp[middle_east_non_emp["Data Source"].isin(countries_list)]
middle_east_countries
```

```python
pd.options.display.max_rows = 60
middle_east_countries 
```

## Cleaning Middle Eastern Non-Employment Data

```python
middle_east_countries = middle_east_countries.iloc[:, [0] + list(range(35, len(middle_east_countries.columns)))]
middle_east_countries
```

```python
middle_east_countries.columns = middle_east_countries.iloc[0 , :]
middle_east_countries = middle_east_countries.reset_index(drop=True)
middle_east_countries =  middle_east_countries.iloc[list(range(1 , len(middle_east_countries))) , :]
pd.set_option('display.max_columns', None)
middle_east_countries.to_excel('data/middle_east_countries.xlsx', index=False)

middle_east_countries
```

```python
middle_east_countries["Country Name"]
```

## Plotting the data

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import LabelEncoder
```

```python
x = np.array(middle_east_countries.columns[range(1 , len(middle_east_countries.columns) )])

y = middle_east_countries.iloc[:, 1:].values.flatten(order = "F")
    
y = np.array(y)
z = np.array(middle_east_countries["Country Name"])

x_coords , z_coords = np.meshgrid(z , x)
x_coords = np.reshape(x_coords , (x_coords.shape[0] * x_coords.shape[1]) )
z_coords = np.reshape(z_coords , (z_coords.shape[0] * z_coords.shape[1]))

from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
lb.fit(x_coords)
x_coords = lb.transform(x_coords)
z_coords= pd.to_numeric(z_coords)

print('x_coords shape:', x_coords.shape)
print('x_coords dtype:', x_coords.dtype)
print('y_coords shape:', y.shape)
print('y_coords dtype:', y.dtype)
print('z_coords shape:', z_coords.shape)
print('z_coords dtype:', z_coords.dtype)
```

```python
figure = plt.figure(figsize=(10, 5) , dpi = 480)
ax = figure.add_subplot(111, projection='3d')
num_segments = 3

surf = ax.plot_trisurf(x_coords,  z_coords , y , linewidth=0.001)

#ax.scatter(x_coords,  z_coords , y )
ax.view_init(azim=-90 , elev = 10)
ax.set_ylabel("Year" , labelpad = 10)
ax.set_zlabel("Unemployment Rate (%)" )
ax.set_xticklabels([])
#ax.set_xlabel("Country index" ,)
figure.suptitle("Unemployment Rate in the Middle East\n(2000-2020)'")
plt.show()
```

```python
import matplotlib.patches as mpatches

figure = plt.figure(figsize=(90, 10) , dpi = 320)
ax = figure.add_subplot(111, projection='3d')
colors = ['red', 'green', 'blue', 'yellow', 'purple', 'orange', 'pink', 'brown', 'gray', 'cyan', 'magenta', 'olive', 'teal', 'navy'] *31
scatter = ax.scatter(x_coords,  z_coords , y  , c = colors)


ax.view_init(azim=0 , elev = 1)
ax.set_ylabel("Year" , labelpad = 10)
ax.set_zlabel("Unemployment Rate (%)" )
ax.set_xticklabels([])
#ax.set_xlabel("Country index" ,)


color_legend = [mpatches.Patch(color=color, label= country) for color,country in zip(colors,z)]
ax.legend(handles=color_legend  ,  bbox_to_anchor=(1.05, 1))

figure.suptitle("Unemployment Rate Trends in the Middle East\n(2000-2020)'")
plt.show()
```
